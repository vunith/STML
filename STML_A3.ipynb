```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
file_path = '/content/drive/MyDrive/housing.csv'
data = pd.read_csv(file_path)
print(data.columns)
columns_of_interest = ['latitude', 'total_rooms', 'housing_median_age','total_bedrooms']
selected_data = data[columns_of_interest]
def calculate_mle(column_data):
  mle_mean = np.mean(column_data)
  mle_variance = np.var(column_data, ddof=0)
  return mle_mean, mle_variance
for column in columns_of_interest:
  mean, variance = calculate_mle(selected_data[column].dropna())
  print(f"Column: {column}")
  print(f" MLE Mean: {mean}")
  print(f" MLE Variance: {variance}")
  print("-" * 30)
```

```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
file_path = '/content/drive/MyDrive/housing.csv'
data = pd.read_csv(file_path)
latitude_data = data['latitude'].dropna()
mean_latitude = np.mean(latitude_data)
variance_latitude = np.var(latitude_data, ddof=0)
def log_likelihood(value, mean, variance):
    return -0.5 * np.log(2 * np.pi * variance) - ((value - mean) ** 2) / (2 *variance)
values_to_check = [50, 75, 80]
for value in values_to_check:
    ll = log_likelihood(value, mean_latitude, variance_latitude)
    print(f"Log-Likelihood for value {value}: {ll}")
print(f"\nMaximum Likelihood Estimate (MLE) for mean: {mean_latitude}")
print(f"Maximum Likelihood Estimate (MLE) for variance: {variance_latitude}")
```

```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
file_path = '/content/drive/MyDrive/housing.csv'
data = pd.read_csv(file_path)
print(data.columns)
columns_of_interest = ['latitude', 'total_rooms', 'housing_median_age','total_bedrooms']
selected_data = data[columns_of_interest]
bins = [5, 10, 15, 20]
for column in columns_of_interest:
  plt.figure(figsize=(8, 5))
  plt.hist(selected_data[column].dropna(), bins=bins, density=True,edgecolor='black', alpha=0.6, color='skyblue') # indented this line
  plt.title(f'Density Estimate for {column}')
  plt.xlabel(column)
  plt.ylabel('Density')
  plt.grid(True)
  plt.show()
```
