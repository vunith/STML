```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
import scipy as sc
import statistics as stat
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, iqr
def find_mode(arr):
  # Indent the lines within the function
  counts = np.bincount(arr)
  mode = np.argmax(counts)
  return mode
# Load the data
data = pd.read_csv("/content/drive/MyDrive/heart_attack_dataset.csv",
                    usecols=['Blood Pressure (mmHg)', 'Cholesterol (mg/dL)'])
print(data.head(20))
# Calculate and display statistics
print("Mean of the data =", np.mean(data))
print("Median of the data =", np.median(data))
print("Mode of the data['Blood Pressure (mmHg)'] =", find_mode(data['Blood Pressure (mmHg)']))
print("Mode of the data['Cholesterol (mg/dL)'] =", find_mode(data['Cholesterol (mg/dL)']))

print("Standard Deviation of the data =", np.std(data))
print("Skew of the data =", skew(data, axis=0, bias=True))
print("Kurtosis of the data =", kurtosis(data, axis=0, bias=True))
print("IQR of the data =", iqr(data))

print("Range of the data =", data.max() - data.min())
# Plot the histograms with different colors
plt.subplot(1, 2, 1)
plt.hist(data['Blood Pressure (mmHg)'], bins=5, color='green',
 edgecolor='black')# Changed color to 'green'
plt.title('Blood Pressure Histogram')
plt.xlabel('Blood Pressure (mmHg)')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
plt.hist(data['Cholesterol (mg/dL)'], bins=5, color='orange',
 edgecolor='black')# Changed color to 'orange'
plt.title('Cholesterol Histogram')
plt.xlabel('Cholesterol (mg/dL)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
