```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
file_path = '/content/drive/MyDrive/archive/train.csv'
data = pd.read_csv(file_path)

data_description = data.describe()
print("Data Description:\n", data_description)
data_info = data.dtypes
data_shape = data.shape
print("\nData Types:\n", data_info)
print("\nData Shape:\n", data_shape)
target_variable = 'price_range'
features = data.drop(target_variable, axis=1)
null_values = data.isnull().sum()
print("\nNull Values:\n", null_values)
data_filled = data.fillna(0)
scaler = MinMaxScaler()
normalized_features = scaler.fit_transform(features)
normalized_features_df = pd.DataFrame(normalized_features, columns=features.
 columns)
print("\nNormalized Features:\n", normalized_features_df)
X_train, X_test, y_train, y_test = train_test_split(normalized_features_df,
 data[target_variable], test_size=0.2, random_state=42)
print("\nTraining Set Shape:", X_train.shape)
print("Testing Set Shape:", X_test.shape)
```
